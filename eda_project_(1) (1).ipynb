```python
#some columns must removed bcs for different reasons(i will mention)removing these columns ensures that the dataset is clean


df.drop (['Soft opening date', 'Fast Lane available', 'Replaced', 'Track layout','Fastrack available',
        'Soft opening date.1','Closing date', 'Opened', 'Replaced by', 'Website', 'Flash Pass Available', 'Must transfer from wheelchair',
        'Theme','latitude', 'longitude', 'Single rider line available', 'Restraint Style', 'Flash Pass available',
    'Acceleration', 'Restraints', 'Name',   # unnecessary bcs they are  mostly empty, r full of na (from column 22 to 41, i mean 19 columns i remove)
    'speed1', 'speed2', 'speed1_value', 'speed1_unit',  # extra speed columns
    'height_unit',  # same with  height_value
          'height_value',# is clean,numeric but in ft format . i want meter format ,so i will clean 'height' column
        'Speed',# i remove it  bcs it is stored as object but speed mph is more clean and nnumeric plus for coastrollers mph is more common than kmh
     'height_ft', # is fullof na
        'Inversions','G-force',# dirty version of Inversions_clean,'G-force_clean
          'opening_date_clean','Opening date',# are same and also i dont need exact date
          'Type'# same with Type_Main                                               so i remove 32 columns
     ],axis=1,inplace=True) # i use inplace for removing these columns from dataset permanently
```

```python
df.info() # i keep just 23 columns for obvious ,clean dataset
```

```python


df['Length'] = pd.to_numeric(df['Length'].str.replace(',', '').str.extract(r'\((\d+)')[0]) # firstly i see differences between writing form like (180 m and 1,014 m )
#when i convert them to numeric 1,014 m be 1. and it is unrealistic(1 m rollercoaster) it is big mistake so,firstly i replace commas,then it is  object
# i must convert it to numeric( meter form) so ,using regex(r'\((\d+)') for pick numbers. '\(' for picking '('      ,     '(\d+)' for 1 or more digits after it

df['Height'] = pd.to_numeric(df['Height'].str.replace(',', '').str.extract(r'\((\d+)')[0])    # same prosses as like lenght

df['Height restriction'] = pd.to_numeric(df['Height restriction'].str.replace(',', '').str.extract(r'\((\d+)')[0]) # same as above.actually i didnt see format differences
#in here but against odds i remove(replace) commas firstly. then use regex . but it is in cm format

# train column is readable and i will not use  it for visualization,also in capasity column i will mention abt count of riders thats way i dont convert it to numeric

df['Capacity'] = pd.to_numeric(df['Capacity'].str.replace(',', '').str.extract(r'(\d+)')[0]) #again i must replace commas first bcs differences like( 1067 n 1,067),
# then keeping only numbers w regex.nnow in capasity column we see just numbers( count of riders for per train ). and in there we dont need '/(' bcs there si no '('

df['Drop'] = pd.to_numeric(df['Drop'].str.replace(',', '').str.extract(r'\((\d+)')[0]) # keeping meter form

# max_vertical_angel,duration are readable in their current version and i will not use them for visualization,so i dont convert it to numeric
```

```python
df.info()  # succesfully convert object to numeric
```

```python
df   #  i again check my changes with looking at dataset
```

```python
df.describe()  # by describing dataset we see that there are outliers,physically impossible numbers
#and we must replace them  with null or none  so our data is safe and makes sense


 #  max lenght is 8133 and this is an unrealistic outlier
              # max height is also unrealistic (1200)
              #  max capacity is again unrealistic (155000)
              # max drop also is outlier(355), we see 50% is 30
               # min height restrictiion is too small(2cm)
               #max speed = outlier
                                     #max  gforce  are in the same boat with others ,commonly it must be max 7
```

```python
df['Length'][(df['Length']>3500)]=None                #  I replaced 3500+ height values with None
df['Height'][(df['Height']>150)]=None
df['Capacity'][(df['Capacity']>2500)]=None
df['Drop'][(df['Drop']>150)]=None
df['Height restriction'][(df['Height restriction']<90)]=None                #  I replaced 90- height restriction  with None
df['speed_mph'][(df['speed_mph']>90)]=None
df['Gforce_clean'][(df['Gforce_clean']>7)]=None

   # in there i do same action for each of them ( replaced outlier with nulls )


```

```python
df.describe()              # so this table is more realistic than previous one
```

```python
df.isna().sum()  # now i have more errors than previous situation(due to converting outliers to nulls)
```

```python
# missing values are handled based on their type
# numerical columns filled with median to keep realistic numbers. median is more reliable than mean
#Categorical columns (objects) filled with mode (most common value) to make sense of missing words
# also for cleaning data we can handle nulls with droping column,raw ( if there are a lot of missing values
```

```python
df['Status'].fillna(df['Status'].mode()[0],inplace=True)  #  use [0] for selects the first (and usually only) mode value bcs without it mode() returns series
df['Length'].fillna(df['Length'].median(),inplace=True)   # as i was mentiooned i fill na with median for numerical data
df['Height restriction'].fillna(df['Height restriction'].median(),inplace=True)
df['Height'].fillna(df['Height'].median(),inplace=True)
df['Capacity'].fillna(df['Capacity'].median(),inplace=True)
df['Drop'].fillna(df['Drop'].median(),inplace=True)
df['speed_mph'].fillna(df['speed_mph'].median(),inplace=True)
df['Inversions_clean'].fillna(df['Inversions_clean'].median(),inplace=True)
df['Gforce_clean'].fillna(df['Gforce_clean'].median(),inplace=True)
df['Manufacturer'].fillna(df['Manufacturer'].mode()[0],inplace=True)
df['Model'].fillna(df['Model'].mode()[0],inplace=True)
df['Lift/launch system'].fillna(df['Lift/launch system'].mode()[0],inplace=True)
df['Trains'].fillna(df['Trains'].mode()[0],inplace=True)
df['Park section'].fillna(df['Park section'].mode()[0],inplace=True)
df['Duration'].fillna(df['Duration'].mode()[0],inplace=True)
df['Designer'].fillna(df['Designer'].mode()[0],inplace=True)
df['Max vertical angle'].fillna(df['Max vertical angle'].mode()[0],inplace=True)
df['Type_Main'].fillna(df['Type_Main'].mode()[0],inplace=True)
df['Cost'].fillna(df['Cost'].mode()[0],inplace=True)

# i do same process for all of them due to same reasons
```

```python
df.isna().sum() # so  dataset is nafree
```

# **visualization** .







```python
#so i willl continue with visualization.it helps Turn data into pictures to
 #understand patterns, trends, or relationships. makes big numbers easy to read.
#helps tell a story with data w bar,line,pie chart or plots
# for my data i will use soome of them step by step.
```

```python
#firstly i wanna compare manufactures through rollercoaster's height
# bcs i want to see top 10 manufacturers abt height
# so,this bar chart shows the top 10 roller coaster manufacturers by average height.
# the tallest coasters are built by the first few manufacturers on the left.
# this tells us which companies design more extreme or high roller coasters.

top10 = (df.groupby('Manufacturer')['Height']
      .mean() # it compute average rollercoaster height for each manufacturer
      .sort_values(ascending=False) #deafult iss ascending we must change it
      .head(10))     # for top 10

plt.figure(figsize=(8,6)) # for giving ur own sizes
top10.plot(kind='bar',color='#C41230') # i want use barchart with red color
plt.xlabel('manufacturer') # naming x label
plt.ylabel('average height (meters)') # naming y label
plt.title('Top 10 manufacturers by average roller coaster height') # naming chart
plt.xticks(rotation=90) # for appropirately pplacement of manufacturer names
plt.show() # display
```

```python
# so,this bar chart shows the top 10 roller coaster manufacturers by average height.
# the tallest coasters are built by the first few manufacturers on the left.
# this tells us which companies design more extreme or high roller coasters.
# as we can obviously see sanoyas hishino meisho produces tallest roller coasters.
```

```python
#Speed distribution

df['speed_mph'].dropna().plot(
    kind='hist', bins=30, figsize=(8, 4), title='Distribution of roller coaster speeds', color='orange'
)
plt.xlabel('Speed (mph)')
plt.show()
```

```python
# Coasters built per year (trend over time)

df.dropna(subset=['year_introduced']).groupby('year_introduced').size().plot(
    kind='line', figsize=(12, 4), title='Number of Roller Coasters Built Per Year'
)
plt.xlabel('Year')
plt.ylabel('Count')
plt.show()
```

```python
#Roller coaster construction peaked in the 2000 era, suggesting a golden age of theme park investment
```

```python
#  i want explore relationshios between height and speed
# bcs it is very interesting for me : height can affect to speed ? taller ones tend to be faster?

plt.figure(figsize=(8,5))
plt.scatter(df['Height'], df['speed_mph'], color='darkgreen')
plt.xlabel('Height (meters)')
plt.ylabel('Speed (mph)')
plt.title('Roller Coaster Height vs Speed')
plt.show()
```
